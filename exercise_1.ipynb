### 1. Programa para sobrepor quadriculado em imagem de cinza.

Escrever um programa que dada uma imagem em níveis de cinza tipo uint8, sobreponha um quadriculado branco (valor 255) de lado definido. A função recebe dois parâmetros: f e delta. O primeiro parâmetro f é a imagem onde o quadriculado será sobreposto e delta é a distância entre as linhas, em coordenadas.

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import os

def sobrepoe_quadriculado(f, delta):
    if os.path.isfile(f):
        im = mpimg.imread(f)
        #cria o array com empty(mais rapido)
        new_im = np.empty(shape=im.shape, dtype='uint8')
        #preenche o array com 255
        new_im.fill(255)
        print(new_im)
        plt.imshow(new_im, cmap="gray")
    else:
        print ('File not found')
```

### 3. Programa para sobrepor quadriculado em imagem de cinza.

Escrever um programa que dada uma imagem em níveis de cinza tipo uint8, sobreponha um quadriculado branco (valor 255) de lado definido. A função recebe dois parâmetros: f e delta. O primeiro parâmetro f é a imagem onde o quadriculado será sobreposto e delta é a distância entre as linhas, em coordenadas.


```python
def rotaciona_90graus(f):
    if os.path.isfile(f):
        im = mpimg.imread(f)
        height, width = im.shape
        new_im = np.zeros(shape=(width, height), dtype='uint8')
        for p in range(width):
            new_im[p] = im[:,p][::-1]    
        plt.imshow(new_im, cmap="gray")
    else:
        print ('File not found')
```

```python
rotaciona_90graus('data/astablet.tif')
```
